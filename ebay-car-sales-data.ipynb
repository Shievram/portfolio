# Exploring price by brand

When working with data on cars, it's natural to explore variations across different car brands. We can use aggregation to understand the brand column.

```python
#Select the first 10 brands
top_brands_counts = autos['brand'].value_counts(normalize=True).head(10)
top_brands = top_brands_counts.index
print(top_brands)
```

```python
#Calculate the mean price of each brand assigned
top_brands_mean_price = {}

for b in top_brands:
    selected_rows = autos[autos['brand'] == b]
    mean_price = selected_rows['price'].astype(int).mean()
    top_brands_mean_price[b] = mean_price
    
print(top_brands_mean_price)
```

# Storing aggregate data in a dataframe

We can combine the data from both series objects into a single dataframe (with a shared index) and display the dataframe directly. To do this, we'll need to learn two pandas methods:

1. pandas series constructor
2. pandas dataframe constructor

```python
#Calculate the mean mileage
top_brands_mean_mileage = {}

for b in top_brands:
    selected_rows = autos[autos["brand"] == b]
    mean_mileage = selected_rows["odometer_km"].astype(int).mean()
    top_brands_mean_mileage[b] = mean_mileage
    
print(top_brands_mean_mileage)
```

```python
#Convert both dictionaries to series constructors
mean_mileage_series = (pd.Series(top_brands_mean_mileage)
                       .sort_values(ascending=True))

print(mean_mileage_series)
```

Fiat displays the lowest mileage whereas BMW is the highest.

```python
mean_price_series = pd.Series(top_brands_mean_price)
print(mean_price_series)
```

Mercedes Benz is the most expensive while Renault is the less expensive

```python
#Convert both dictionaries to dataframe constructors
brand_dataframe = pd.DataFrame(mean_mileage_series, columns=["mean_mileage"])
brand_dataframe
```

```python
brand_dataframe["mean_price"] = mean_price_series
brand_dataframe
```
